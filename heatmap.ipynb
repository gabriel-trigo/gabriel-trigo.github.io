```python
import tensorflow as tf
import seaborn as sns
import numpy as np
from qlearning_agent import QlearningAgent
from game import game_instance
```

```python
model = tf.keras.models.load_model("trial_jun-27-2/model_9/model")
```

```python
agent = QlearningAgent(eps_start=0)
```

```python
agent.nn.model = model
```

```python
game_instance(agent=agent, render=True)
```

```python
pipe_height = 100
states = []
for y in range(500, 0, -20):
    for x in range(500, 0, -20):
        states.append([y - pipe_height - 100, 0, x, y - pipe_height + 100])
```

```python
eval = agent.nn.model.predict(states)
```

```python
values = np.array([max(el) for el in eval])
```
